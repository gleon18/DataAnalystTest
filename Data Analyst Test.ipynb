```python
import pandas as pd
from datetime import datetime
import numpy as np
from statsmodels.formula.api import ols
```

### Initially I just review the data, change the variable review_time to a Date Type, check if the data has some null rows and describe the data set to understand the scale of the review variables

```python
df = pd.read_csv("beer_reviews.csv", delimiter=',')
df
```

```python
df.info()
```

```python
# Using Describe function to check the scale of the review variables
df.describe()
```

```python
df['review_time'] = pd.to_datetime(df['review_time'], unit='s')
df['review_time'] = df['review_time'].fillna(0)
df.dropna(subset=['review_time'], inplace=True)
```

```python
null_counts = df.isnull().sum(axis=0)
print(null_counts)
```

```python
# I cleaned the nulls from the beer_abv field and then calculate the mean of abv by Brewery
StrongestBeer=df
StrongestBeer['beer_abv']=StrongestBeer['beer_abv'].fillna(0)
StrongestBeer.dropna(subset=['beer_abv'], inplace=True)
null_counts = df.isnull().sum(axis=0)
print(null_counts)
StrongestBeer = df.groupby(['brewery_name','beer_name'])['beer_abv'].mean()
StrongestBeer  = StrongestBeer .reset_index()
StrongestBeer  = pd.DataFrame(StrongestBeer)
pivot = StrongestBeer.pivot_table(index=['brewery_name'], values=['beer_abv'], aggfunc='mean')
StrongestBeer = pd.DataFrame(pivot)
StrongestBeer = StrongestBeer.sort_values(by='beer_abv',ascending=False)
StrongestBeer.head(10)
```

## Which brewery produces the strongest beers by abv ? 

   **Schorschbräu** with a mean by beer of *24,7*

```python
#Create avariable that takes into account the 5 aspects and then get the mean of this variable by beer and the number of reviews each beer got 
df['BestBeerMean'] = df.apply(lambda row: (row['review_overall'] + row['review_aroma']+row['review_appearance']+row['review_palate']+row['review_taste'])/5, axis=1)
BestBeer = df.groupby(['beer_name']).agg({'BestBeerMean':'mean', 'beer_beerid':'size'})
BestBeer= BestBeer .reset_index()
BestBeer  = pd.DataFrame(BestBeer)
BestBeer= BestBeer.sort_values(by=['beer_beerid', 'BestBeerMean'], ascending=False)
BestBeer.head(10)

```

## If you had to pick 3 beers to recommend to someone, how would you approach the problem ? 
I would get the mean of the 5 aspects and  I will count the number of reviews to select the top 3 with both the highher number of reviews with high Score mean , in this case I would recommend **90 Minute IPA,India Pale Ale	and Old Rasputin Russian Imperial Stout**

### To get the factors that impact the most the Quality of the beer, I calculate the correlation coefficient of the different factors with the Overall impression variable

```python
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
#Linear Regression
formula = 'review_overall ~ review_aroma + review_appearance + review_palate + review_taste'
model = ols(formula, data=df).fit()
print(model.summary())
```

## What are the factors that impacts the quality of beer the most ?
The higher correlation coeficient for the different factors and the Overall Rating is the **Taste** and also when trying to fit a Regression Model is the variable with the highest coefficient 

```python
df['beer_style'].describe()
```

```python
# Getting the most consistent Beer Style in Aroma and Appereance
BestStyle = df.groupby(['beer_style'])['review_aroma','review_appearance'].mean()
BestStyle= BestStyle .reset_index()
BestStyle  = pd.DataFrame(BestStyle)
BestStyle['AAMean']=BestStyle.apply(lambda row: ( row['review_aroma']+row['review_appearance'])/2, axis=1)
BestStyle = BestStyle.sort_values(by='AAMean',ascending=False)
BestStyle.head(3)
```

```python
#Getting the beer within that Beer Style with the highest approval and higher number of reviews
BeerByStyle = df.groupby(['beer_style','beer_name']).agg({'BestBeerMean':'mean', 'beer_beerid':'size'})
BeerByStyle= BeerByStyle .reset_index()
BeerByStyle  = pd.DataFrame(BeerByStyle)
BeerByStyle=BeerByStyle.loc[BeerByStyle['beer_style'] == 'American Double / Imperial Stout']
BestBeer= BeerByStyle.sort_values(by=['beer_beerid', 'BestBeerMean'], ascending=False)
BestBeer.head(10)
```
